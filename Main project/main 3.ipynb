```python
df1 = pd.read_csv("Resume.csv")
df = df1.drop(df1.iloc[:,4:],axis = 1)
df = df.reindex(np.random.permutation(df.index))
data = df.copy().iloc[
    0:200,
]
data.head()
```

```python
nlp = spacy.load("en_core_web_lg")
skill_pattern_path = "jz_skill_patterns.jsonl"
```

```python
ruler = nlp.add_pipe("entity_ruler")
ruler.from_disk(skill_pattern_path)
nlp.pipe_names
```

```python
def get_skills(text):
    doc = nlp(text)
    myset = []
    subset = []
    for ent in doc.ents:
        if ent.label_ == "SKILL":
            subset.append(ent.text)
    myset.append(subset)
    return subset


def unique_skills(x):
    return list(set(x))
```

```python
clean = []
for i in range(data.shape[0]):
    review = data["Resume_str"].iloc[i]

    # Check if review is a string
    if isinstance(review, str):
        # Clean the text using regular expressions
        review = re.sub('(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)|^rt|http.+?', ' ', review)
        review = review.lower()
        review = review.split()

        # Lemmatize and remove stopwords
        lm = WordNetLemmatizer()
        review = [lm.lemmatize(word) for word in review if word not in set(stopwords.words("english"))]

        review = " ".join(review)
        clean.append(review)
    else:
        # Handle the case where the value isn't a string
        clean.append("")  # Or you can choose another approach to handle non-string values
```

```python
data["Clean_Resume"] = clean
data["skills"] = data["Clean_Resume"].str.lower().apply(get_skills)
data["skills"] = data["skills"].apply(unique_skills)
data.head()
```

```python
docs = data["Clean_Resume"].values
dictionary = corpora.Dictionary(d.split() for d in docs)
bow = [dictionary.doc2bow(d.split()) for d in docs]
lda = gensim.models.ldamodel.LdaModel
num_topics = 4
ldamodel = lda(
    bow,
    num_topics=num_topics,
    id2word=dictionary,
    passes=50,
    minimum_probability=0
)
ldamodel.print_topics(num_topics=num_topics)
```

```python
pyLDAvis.enable_notebook()
pyLDAvis.gensim_models.prepare(ldamodel, bow, dictionary)
```
